*1. what is the difference between a function and a method in Python?*

```python
# function - A function is a block of code that performs a specific task and can be called anywhere in the code.
# Method - Methods are functions that are defined inside a class and are called on objects. It is always associated with an object.
```

```python
# function
def add(a,b):
  return a+b

add(3,4)
```

```python
#Method
class Calc:
  def add(self, a, b):
    return a+b

c = Calc()
print(c.add(3 ,4))
```

*3. What are the different ways to define and call a function in python.*

```python
#Different ways to define and calling a functions:-
#Regular function - Defining with def and call by name.
#Lambda function - Defining by using lambda keyword and it is anonymous.
#generator function - It used yield to return values one at a time.
```

```python
#Regular Function
def add(a,b):
  return a+b

add(3,4)
```

```python
# Lambda Function
a = (lambda x,y:x*y)
a(5,6)
```

```python
# Generator function
def square(n):
  l = []
  for i in n:
    l.append(i**2)
    yield l



a = square([1,2,3])
```

```python
next(a)
```

*6. Explain the concept of generators in Python and how they are defined.*

```python
# Generators are a simple way to create iterators using functions and the yield keyword.
# It produce values one at a time and only when required.
# It takes less memory usuage and time for output.
```

```python
def add(n):
  for i in range(n):
    yield i+10
```

```python
a = add(5)
next(a)
```

```python
next(a)
```

*7. What are the advantages of using generators over regurlar function*

```python
# Advantages of using generators over regular function are:-
# Memory Efficiency - Generators yield one at a time, meaning they don't store the entire result in memory.
# Less Time - It take less time for large data because it give value one at a time.
```

```python
def cube(a):
  l = []
  for i in a:
    l.append(i**3)
    yield l
```

```python
l = [1,5,10,15]
y = cube(l)
y
```

```python
next(y)
```

*10. What is the difference between map(), reduce and filter() functions in python?*

```python
# Map - Applies a function to each item in a list and returns a new list with the transformed items.
# Filter - Select items from list that meet a given condition.
# Reduce - Combine all items in a list into a single value by applying a function cumulatively.
```

```python
# Map
numbers = [1,2,3,4,5,6,7]
w = map(lambda x : x**3, numbers)
print(list(w))
```

```python
# Filter
numbers = [1,2,3,4,5,6,7]
e = filter(lambda x:x%2==0,numbers)
print(list(e))
```

```python
from functools import reduce
```

```python
# Reduce
numbers = [1,2,3,4,5,6,7]
s = reduce(lambda x,y: x+y, numbers)
print(s)
```
